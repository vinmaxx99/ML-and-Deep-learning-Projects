```python
#Evaluating Regression Models
import numpy as nm
import matplotlib.pyplot as plt
```

```python
x=nm.linspace(1,4,1000)
y=x**2+3*x+5
y=y+nm.random.randn(1000)
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
plt.scatter(x,y)
```

# Linear Regression



```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train.reshape(-1,1),y_train.reshape(-1,1))

#Visualizing Result
y_pred=regressor.predict(x_test.reshape(-1,1))

plt.scatter(x_train,y_train,color='red')
plt.plot(x_train.reshape(-1,1),regressor.predict(x_train.reshape(-1,1)))
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_test.reshape(-1,1),y_pred)
```

# Polynomial Regression

```python
#Fitting Polynomial
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=2)
X_poly=poly_reg.fit_transform(x.reshape(-1,1))

from sklearn.cross_validation import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_poly,y,test_size=0.2)

lin_reg=LinearRegression()
lin_reg.fit(X_train,Y_train.reshape(-1,1))
#lin_reg.predict(X_train)
```

```python
plt.scatter(X_test[:,1],Y_test)
plt.plot(X_test[:,1],lin_reg.predict(X_test),color='red')
#plt.xlim(1,2)
#plt.ylim(1,15)
#Some Doubts
```

# SVM Regression

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
x_train=sc_X.fit_transform(x_train.reshape(-1,1))
x_test=sc_X.transform(x_test.reshape(-1,1))
sc_Y=StandardScaler()
y_train=sc_Y.fit_transform(y_train.reshape(-1,1))
y_test=sc_Y.transform(y_test.reshape(-1,1))

from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
regressor.fit(x_train,y_train)


```

```python
plt.plot(x_train,regressor.predict(x_train))
plt.scatter(x_train,y_train,color='red')
```

# Decision Tree

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train.reshape(-1,1),y_train.reshape(-1,1))

y_pred=regressor.predict(x_test.reshape(-1,1))
```

```python
#Plot
plt.scatter(x_test,y_test)
plt.scatter(x_test,y_pred,color='red')
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=10)
regressor.fit(x_train.reshape(-1,1),y_train.reshape(-1,1))
y_pred=regressor.predict(x_test.reshape(-1,1))
```

```python
plt.scatter(x_test,y_test)
plt.scatter(x_test,y_pred,color='red')
```
