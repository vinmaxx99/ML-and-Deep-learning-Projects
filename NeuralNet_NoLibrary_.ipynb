```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

###Procedure
1. Make Data
2. Make random weight
3. Define Activations
4. for each epoch update synapses

```python
#Data Creation
X1=np.linspace(0,100,100)
X1=np.repeat(X1,100)
X2=np.linspace(0,100,100)
X2=np.tile(X2,100)
Y=X1**2+X2**2
y=[]
for i in Y:
    if(i>50**2 and i<80**2):
        y.append(1)
    else:
        y.append(0)


Y=np.array(y)
```

```python
#Visualization
plt.plot(X1[Y<1],X2[Y<1],color='red')
plt.plot(X1[Y>0],X2[Y>0],color='blue')
```

```python
#Training Neural Network
no_of_epoch=1000
noh_layer1=4
noh_layer2=4
W1=np.random.randn(4,2)
W2=np.random.randn(4,4)
W3=np.random.randn(4,1)
X=np.vstack((X1,X2))
#Activation='sigmoid'

def sigmoid(x):
    return 1/1+np.exp(x)
def sigmoidrev(x):
    return x*(1-x)
def relu(x):
    return x*(x>0)
def relurev(x):
    x[x<=0] = 0
    x[x>0] = 1
    return x
```

```python
for i in range(no_of_epoch):
    #Forward Prop
    l1=sigmoid(W1.dot(X))
    l2=sigmoid(W2.dot(l1))
    l3=sigmoid(W3.T.dot(l2))
    error_l3=Y-l3
    if(i%100==0):
        print("Error: "+str(np.mean(np.abs(error_l3))))
    #Error Calculation and Backprop
    delta_l3=error_l3*sigmoidrev(l3)
    error_l2=W3.dot(delta_l3)
    delta_l2=error_l2*sigmoidrev(l2)
    error_l1=W2.dot(delta_l2)
    delta_l1=error_l1*sigmoidrev(l1)
    #Updating Synapses
    W1+=delta_l1.dot(X.T)
    W2+=delta_l2.dot(l1.T)
    get=delta_l3.dot(l2.T)
    W3+=get.T
```
